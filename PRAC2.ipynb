```python
import matplotlib.pyplot as plt
import random

def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            break
    return comparisons

input_sizes = [100, 500, 1000, 5000, 10000]
best_case = []
average_case = []
worst_case = []

for size in input_sizes:
    arr = list(range(1, size + 1))


    best_case.append(linear_search(arr, arr[0]))


    average_case.append(linear_search(arr, arr[size // 2]))

    worst_case.append(linear_search(arr, -1))


plt.plot(input_sizes, best_case, marker='o', label='Best Case (O(1))')
plt.plot(input_sizes, average_case, marker='o', label='Average Case (O(n/2))')
plt.plot(input_sizes, worst_case, marker='o', label='Worst Case (O(n))')

plt.title("Linear Search: Number of Comparisons vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid(True)
plt.show()
```
