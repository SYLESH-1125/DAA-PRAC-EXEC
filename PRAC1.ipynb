```python
import random
import time
import matplotlib.pyplot as plt

# Linear search implementation
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate random sequence
def generate_random_sequence(length):
    return random.sample(range(1, length + 1), length)

# Plotting function
def plot_time_complexity(x, y_best, y_avg, y_worst):
    plt.plot(x, y_best, marker='o', label='Best Case (Start)', color='green')
    plt.plot(x, y_avg, marker='o', label='Average Case (Middle)', color='blue')
    plt.plot(x, y_worst, marker='o', label='Worst Case (Not Present)', color='red')
    plt.xlabel('Sequence Length')
    plt.ylabel('Average Time (seconds)')
    plt.title('Time Complexity of Linear Search (Best vs Average vs Worst)')
    plt.legend()
    plt.show()

# Parameters
sequence_lengths = [5000, 10000, 15000, 20000]
num_trials = 100

average_times = {
    'Best': [],
    'Average': [],
    'Worst': []
}

# Main loop for timing
for length in sequence_lengths:
    total_time_best = 0
    total_time_avg = 0
    total_time_worst = 0

    for _ in range(num_trials):
        sequence = generate_random_sequence(length)

        val_best = sequence[0]               # Best case: first element
        val_avg = sequence[length // 2]      # Average case: middle element
        val_worst = -1                       # Worst case: not present

        # Best case
        start = time.time()
        linear_search(sequence, val_best)
        total_time_best += (time.time() - start)

        # Average case
        start = time.time()
        linear_search(sequence, val_avg)
        total_time_avg += (time.time() - start)

        # Worst case
        start = time.time()
        linear_search(sequence, val_worst)
        total_time_worst += (time.time() - start)

    average_times['Best'].append(total_time_best / num_trials)
    average_times['Average'].append(total_time_avg / num_trials)
    average_times['Worst'].append(total_time_worst / num_trials)

# Plotting
plot_time_complexity(sequence_lengths, average_times['Best'], average_times['Average'], average_times['Worst'])
```
